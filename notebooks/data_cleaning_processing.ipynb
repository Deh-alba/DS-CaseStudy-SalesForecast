FIll the Nan sell_price with the Median. Here we use median instead average because is resitant to outliers

```python
# Step 1: Calculate the median price for each combination of store_id and item_id
median_prices = sales_prices.groupby(['store_id', 'item_id'])['sell_price'].median().reset_index()
median_prices.rename(columns={'sell_price': 'median_price'}, inplace=True)
```

```python
# Step 2: Merge the median prices back into the sales_long DataFrame
sales_long = sales_long.merge(median_prices, on=['store_id', 'item_id'], how='left')
```

```python
# Step 3: Fill missing sell_price values with the median price
sales_long['sell_price'].fillna(sales_long['median_price'], inplace=True)
```

```python
# Step 4: Drop the median_price column (optional, if no longer needed)
sales_long.drop(columns=['median_price'], inplace=True)
```

```python
sales_long["date"] = pd.to_datetime(sales_long["date"])
```

```python
# Time Features
sales_long["day"] = sales_long["date"].dt.day
sales_long["month"] = sales_long["date"].dt.month
sales_long["quarter"] = sales_long["date"].dt.quarter
sales_long["day_of_week"] = sales_long["date"].dt.weekday
sales_long["is_weekend"] = (sales_long["day_of_week"] >= 5).astype(int)
sales_long["is_holiday"] = sales_long["event_name_1"].notnull().astype(int)
```

```python
sales_long.head()
```

```python
print(sales_long["sales"].isnull().sum())  # Check if there are missing values in the sales column
```

```python
# Create lag features
for lag in [7, 14, 28]:
    sales_long[f"lag_{lag}"] = sales_long.groupby(["store_id","item_id"])["sales"].shift(lag)

# Fill NaN values in lag features with the median sales for each item_id
for lag in [7, 14, 28]:
    sales_long[f"lag_{lag}"] = sales_long.groupby("item_id")[f"lag_{lag}"].transform(
        lambda x: x.bfill().fillna(x.rolling(window=3, min_periods=1).median())
    )

# Fill any remaining NaN values with 0 as a fallback
for lag in [7, 14, 28]:
    sales_long[f"lag_{lag}"].fillna(-999, inplace=True)
```

```python
sales_long.head(30)
```

```python
# Rolling Mean (Trend Features)
for window in [7, 14, 28]:
    sales_long[f"rolling_mean_{window}"] = sales_long.groupby(["store_id","item_id"])["sales"].transform(
        lambda x: x.shift(7).rolling(window).mean()
    )
    
for window in [7, 14, 28]:
    sales_long[f"rolling_mean_{window}"] = sales_long.groupby("item_id")[f"rolling_mean_{window}"].transform(
        lambda x: x.bfill().fillna(x.rolling(window=3, min_periods=1).median())
    )
    

for lag in [7, 14, 28]:
    sales_long[f"rolling_mean_{window}"].fillna(-999, inplace=True)
```

```python
sales_long["price_lag_7"] = sales_long.groupby(["store_id","item_id"])["sell_price"].shift(7)


# Fill NaN values in price_lag_7 with the median sell_price for each store_id and item_id
sales_long["price_lag_7"] = sales_long.groupby(["store_id","item_id"])["price_lag_7"].transform(
    lambda x: x.bfill().fillna(x.rolling(window=3, min_periods=1).median())
)

# Recalculate price_change
sales_long["price_change"] = (sales_long["sell_price"] - sales_long["price_lag_7"]) / sales_long["price_lag_7"]


# Fill any remaining NaN values in price_change with 0 (e.g., no change)
sales_long["price_change"].fillna(-999, inplace=True)
```
